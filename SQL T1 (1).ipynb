```python
#2 to create database intern_training_db
conn = sqlite3.connect("intern_training_db.db")
cursor = conn.cursor()
```

```python
#4
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    email TEXT,
    age INTEGER
)
""")
```

```python
#2
cursor.execute("""
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    email VARCHAR(100) UNIQUE NOT NULL,
    age INT NOT NULL,
    dob DATE
)
""")
conn.commit()
```

```python
#3
valid_students = [
    (1, "Rahul Sharma", "rahul@gmail.com", 21, "2003-05-10"),
    (2, "Ananya Verma", "ananya@gmail.com", 22, "2002-08-15"),
    (3, "Karan Mehta", "karan@gmail.com", 20, "2004-01-20")
]

cursor.executemany("""
INSERT INTO students (id, name, email, age, dob)
VALUES (?, ?, ?, ?, ?)
""", valid_students)

conn.commit()
```

```python
#7
cursor.execute("""
CREATE TABLE students_new (
    id INTEGER PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    email VARCHAR(100) UNIQUE NOT NULL,
    age INT NOT NULL,
    dob DATE
)
""")

cursor.execute("""
INSERT INTO students_new (id, name, email, age, dob)
SELECT id, name, email, age, dob FROM students
""")

cursor.execute("DROP TABLE students")
cursor.execute("ALTER TABLE students_new RENAME TO students")
conn.commit()
```

```python
#8
#PRIMARY KEY → Unique student identification
#NOT NULL → Mandatory core attributes
#UNIQUE → No duplicate emails
#Proper types → Reliable analytics & joins
```
